```python
police_name = []
for name in data['관서명']:
    police_name.append('서울' + str(name[:-1]) + '경찰서')
    
police_name
```

```python
import folium
from tqdm import tqdm_notebook

police_address = []
police_lat = []
police_lng = []
for name in tqdm_notebook(police_name):
    tmp = gmaps.geocode(name, language='ko')
    police_address.append(tmp[0].get("formatted_address"))
    tmp2 = tmp[0].get('geometry')
    police_lat.append(tmp2['location']['lat'])
    police_lng.append(tmp2['location']['lng'])
    print(police_address, police_lat, police_lng)
    
len(police_address), len(police_lat), len(police_lng)
```

```python
df = pd.DataFrame({'address' : police_address, 'lat' : police_lat, 'lng' : police_lng})
df.head()
```

```python
mapping = folium.Map(location=[df['lat'].mean(),df['lng'].mean()], zoom_start=11)

for n in df.index:
    
    folium.Marker([df['lat'][n],df['lng'][n]], popup=df['address'][n]).add_to(mapping)
    
mapping
```
