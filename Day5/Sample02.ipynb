```python
import numpy as np
from matplotlib import pyplot as plt
x = np.arange(1, 11)
print(x)
y = 2 * x + 5
print(y)
plt.title("Test Plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x, y,"D")
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.sin(x)
plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.show()
```
